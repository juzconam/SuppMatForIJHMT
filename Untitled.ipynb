```python
for i in range(col): # 对列进行遍历
    cols = np.matrix(table.col_values(i)) # 把list转换为矩阵进行矩阵操作
    datamatrix[:, i] = cols # 按列把数据存进矩阵中
```

```python
train = np.zeros((391,27)) 
train[:,0]   = datamatrix[:,0]*1e3      # Pa
train[:,1]   = datamatrix[:,1]          # K
train[:,2]   = datamatrix[:,2]/1000     # Dh [m]
train[:,3]   = datamatrix[:,3]          # °
train[:,4]   = datamatrix[:,4]*1e4      # w/m2
train[:,5]   = datamatrix[:,5]          # m/s
```

```python
T_sat        = datamatrix[:,6]
D            = datamatrix[:,7] # 密度
D_vapor      = datamatrix[:,8] # 气体密度
D_liquid     = datamatrix[:,9] # 液体密度
Surf_tension = datamatrix[:,10] # 表面张力
Cp           = datamatrix[:,11] # 密度
Cp_vapor     = datamatrix[:,12] # 气体密度
Cp_liquid    = datamatrix[:,13] # 液体密度
V_vapor      = datamatrix[:,14] # 气体动力粘度
V_liquid     = datamatrix[:,15] # 液体动力粘度
V            = datamatrix[:,16] # 密度
Pr_vapor     = datamatrix[:,17] # 气体Pr
Pr_liquid    = datamatrix[:,18] # 液体Pr
Pr           = datamatrix[:,19] # 密度 
```

```python
G            = D * train[:,5] #kg/m2s
Bd           = 9.8 * (D_liquid -D_vapor) * train[:,2] / Surf_tension # Bond :ber
Fr_vapor     = G ** 2 / (D_vapor ** 2 *9.8 * train[:,2]) # Froude :ber
Fr_liquid    = G ** 2 / (D_liquid ** 2 *9.8 * train[:,2])
Fr           = G ** 2 / (D ** 2 *9.8 * train[:,2])
Ga           = D_liquid * 9.8 * (D_liquid - D_vapor) * train[:,2] ** 3 / V_liquid ** 2 # Galileo :ber
Ka           = V_liquid ** 4 * 9.8 / D_liquid * Surf_tension ** 3 
Re_vapor     = G * train[:,2] / V_vapor
Re_liquid    = G * train[:,2] / V_liquid
Re           = G * train[:,2] / V
Pe_vapor     = Re_vapor * Pr_vapor   # vapor Peclet :ber 
Pe_liquid    = Re_liquid * Pr_liquid # liquid Peclet :ber 
Pe           = Re * Pr
Su_vapor     = Surf_tension * D_vapor * train[:,2] / V_vapor ** 2 # vapor Suratman :ber 
Su_liquid    = Surf_tension * D_liquid * train[:,2] / V_liquid ** 2 # liquid Suratman :ber
Su           = Surf_tension * D * train[:,2] / V ** 2
We_vapor     = G ** 2 * train[:,2] / D_vapor / Surf_tension # vapor Weber :ber 
We_liquid    = G ** 2 * train[:,2] / D_liquid / Surf_tension  # liquid Weber :ber
We           = G ** 2 * train[:,2] / D / Surf_tension 
        
train[:,6]   = Bd
train[:,7]   = Fr_vapor  
train[:,8]   = Fr_liquid  
train[:,9]   = Fr 
train[:,10]  = Ga
train[:,11]  = Pr_vapor
train[:,12]  = Pr_liquid
train[:,13]  = Pr
train[:,14]  = Ka
train[:,15]  = Re_vapor
train[:,16]  = Re_liquid
train[:,17]  = Re
train[:,18]  = Pe_vapor
train[:,19]  = Pe_liquid
train[:,20]  = Pe
train[:,21]  = Su_vapor
train[:,22]  = Su_liquid
train[:,23]  = Su
train[:,24]  = We_vapor
train[:,25]  = We_liquid
train[:,26]  = We
print(train.shape)
datta = np.zeros((391,15)) 
datta[:,0]   = datamatrix[:,0]*1e3      # Pa
datta[:,1]   = datamatrix[:,1]          # K
datta[:,2]   = datamatrix[:,2]/1000     # Dh [m]
datta[:,3]   = datamatrix[:,3]          # °
datta[:,4]   = datamatrix[:,4]*1e4      # w/m2
datta[:,5]   = datamatrix[:,5]          # m/s
datta[:,6]   = Bd
datta[:,7]   = Fr 
datta[:,8]  = Ga
datta[:,9]  = Pr
datta[:,10]  = Ka
datta[:,11]  = Re
datta[:,12]  = Pe
datta[:,13]  = Su
datta[:,14]  = We
```

```python
def adj_r_squared(x_test,y_test,y_predict):
    SS_R = sum((y_test-y_predict)**2)
    SS_T = sum((y_test-np.mean(y_test))**2)
    r_squared = 1 - (float(SS_R))/SS_T
    adj_r_squared = 1 - (1-r_squared)*(len(y_test)-1)/(len(y_test)-x_test.shape[1]-1)
    return adj_r_squared
```

```python
train2 = np.zeros((94,27)) 
train2[:,0]   = datamatrix2[:,0]*1e3      # Pa
train2[:,1]   = datamatrix2[:,1]          # K
train2[:,2]   = datamatrix2[:,2]/1000     # Dh [m]
train2[:,3]   = datamatrix2[:,3]          # °
train2[:,4]   = datamatrix2[:,4]*1e4      # w/m2
train2[:,5]   = datamatrix2[:,5]          # m/s
```

```python
T_sat        = datamatrix2[:,6]
D            = datamatrix2[:,7] # 密度
D_vapor      = datamatrix2[:,8] # 气体密度
D_liquid     = datamatrix2[:,9] # 液体密度
Surf_tension = datamatrix2[:,10] # 表面张力
Cp           = datamatrix2[:,11] # 密度
Cp_vapor     = datamatrix2[:,12] # 气体密度
Cp_liquid    = datamatrix2[:,13] # 液体密度
V_vapor      = datamatrix2[:,14] # 气体动力粘度
V_liquid     = datamatrix2[:,15] # 液体动力粘度
V            = datamatrix2[:,16] # 密度
Pr_vapor     = datamatrix2[:,17] # 气体Pr
Pr_liquid    = datamatrix2[:,18] # 液体Pr
Pr           = datamatrix2[:,19] # 密度 
```

```python
G            = D * train2[:,5] #kg/m2s
Bd           = 9.8 * (D_liquid -D_vapor) * train2[:,2] / Surf_tension # Bond :ber
Fr_vapor     = G ** 2 / (D_vapor ** 2 *9.8 * train2[:,2]) # Froude :ber
Fr_liquid    = G ** 2 / (D_liquid ** 2 *9.8 * train2[:,2])
Fr           = G ** 2 / (D ** 2 *9.8 * train2[:,2])
Ga           = D_liquid * 9.8 * (D_liquid - D_vapor) * train2[:,2] ** 3 / V_liquid ** 2 # Galileo :ber
Ka           = V_liquid ** 4 * 9.8 / D_liquid * Surf_tension ** 3 
Re_vapor     = G * train2[:,2] / V_vapor
Re_liquid    = G * train2[:,2] / V_liquid
Re           = G * train2[:,2] / V
Pe_vapor     = Re_vapor * Pr_vapor   # vapor Peclet :ber 
Pe_liquid    = Re_liquid * Pr_liquid # liquid Peclet :ber 
Pe           = Re * Pr
Su_vapor     = Surf_tension * D_vapor * train2[:,2] / V_vapor ** 2 # vapor Suratman :ber 
Su_liquid    = Surf_tension * D_liquid * train2[:,2] / V_liquid ** 2 # liquid Suratman :ber
Su           = Surf_tension * D * train2[:,2] / V ** 2
We_vapor     = G ** 2 * train2[:,2] / D_vapor / Surf_tension # vapor Weber :ber 
We_liquid    = G ** 2 * train2[:,2] / D_liquid / Surf_tension  # liquid Weber :ber
We           = G ** 2 * train2[:,2] / D / Surf_tension 
        
train2[:,6]   = Bd
train2[:,7]   = Fr_vapor  
train2[:,8]   = Fr_liquid  
train2[:,9]   = Fr 
train2[:,10]  = Ga
train2[:,11]  = Pr_vapor
train2[:,12]  = Pr_liquid
train2[:,13]  = Pr
train2[:,14]  = Ka
train2[:,15]  = Re_vapor
train2[:,16]  = Re_liquid
train2[:,17]  = Re
train2[:,18]  = Pe_vapor
train2[:,19]  = Pe_liquid
train2[:,20]  = Pe
train2[:,21]  = Su_vapor
train2[:,22]  = Su_liquid
train2[:,23]  = Su
train2[:,24]  = We_vapor
train2[:,25]  = We_liquid
train2[:,26]  = We
print(train2.shape)
datta2 = np.zeros((94,15)) 
datta2[:,0]   = datamatrix2[:,0]*1e3      # Pa
datta2[:,1]   = datamatrix2[:,1]          # K
datta2[:,2]   = datamatrix2[:,2]/1000     # Dh [m]
datta2[:,3]   = datamatrix2[:,3]          # °
datta2[:,4]   = datamatrix2[:,4]*1e4      # w/m2
datta2[:,5]   = datamatrix2[:,5]          # m/s
datta2[:,6]   = Bd
datta2[:,7]   = Fr 
datta2[:,8]  = Ga
datta2[:,9]  = Pr
datta2[:,10]  = Ka
datta2[:,11]  = Re
datta2[:,12]  = Pe
datta2[:,13]  = Su
datta2[:,14]  = We
```

```python
train3 = np.zeros((76,27)) 
train3[:,0]   = datamatrix3[:,0]*1e3      # Pa
train3[:,1]   = datamatrix3[:,1]          # K
train3[:,2]   = datamatrix3[:,2]/1000     # Dh [m]
train3[:,3]   = datamatrix3[:,3]          # °
train3[:,4]   = datamatrix3[:,4]*1e4      # w/m2
train3[:,5]   = datamatrix3[:,5]          # m/s
T_sat        = datamatrix3[:,6]
D            = datamatrix3[:,7] # 密度
D_vapor      = datamatrix3[:,8] # 气体密度
D_liquid     = datamatrix3[:,9] # 液体密度
Surf_tension = datamatrix3[:,10] # 表面张力
Cp           = datamatrix3[:,11] # 密度
Cp_vapor     = datamatrix3[:,12] # 气体密度
Cp_liquid    = datamatrix3[:,13] # 液体密度
V_vapor      = datamatrix3[:,14] # 气体动力粘度
V_liquid     = datamatrix3[:,15] # 液体动力粘度
V            = datamatrix3[:,16] # 密度
Pr_vapor     = datamatrix3[:,17] # 气体Pr
Pr_liquid    = datamatrix3[:,18] # 液体Pr
Pr           = datamatrix3[:,19] # 密度 
G            = D * train3[:,5] #kg/m2s
Bd           = 9.8 * (D_liquid -D_vapor) * train3[:,2] / Surf_tension # Bond :ber
Fr_vapor     = G ** 2 / (D_vapor ** 2 *9.8 * train3[:,2]) # Froude :ber
Fr_liquid    = G ** 2 / (D_liquid ** 2 *9.8 * train3[:,2])
Fr           = G ** 2 / (D ** 2 *9.8 * train3[:,2])
Ga           = D_liquid * 9.8 * (D_liquid - D_vapor) * train3[:,2] ** 3 / V_liquid ** 2 # Galileo :ber
Ka           = V_liquid ** 4 * 9.8 / D_liquid * Surf_tension ** 3 
Re_vapor     = G * train3[:,2] / V_vapor
Re_liquid    = G * train3[:,2] / V_liquid
Re           = G * train3[:,2] / V
Pe_vapor     = Re_vapor * Pr_vapor   # vapor Peclet :ber 
Pe_liquid    = Re_liquid * Pr_liquid # liquid Peclet :ber 
Pe           = Re * Pr
Su_vapor     = Surf_tension * D_vapor * train3[:,2] / V_vapor ** 2 # vapor Suratman :ber 
Su_liquid    = Surf_tension * D_liquid * train3[:,2] / V_liquid ** 2 # liquid Suratman :ber
Su           = Surf_tension * D * train3[:,2] / V ** 2
We_vapor     = G ** 2 * train3[:,2] / D_vapor / Surf_tension # vapor Weber :ber 
We_liquid    = G ** 2 * train3[:,2] / D_liquid / Surf_tension  # liquid Weber :ber
We           = G ** 2 * train3[:,2] / D / Surf_tension 
        
train3[:,6]   = Bd
train3[:,7]   = Fr_vapor  
train3[:,8]   = Fr_liquid  
train3[:,9]   = Fr 
train3[:,10]  = Ga
train3[:,11]  = Pr_vapor
train3[:,12]  = Pr_liquid
train3[:,13]  = Pr
train3[:,14]  = Ka
train3[:,15]  = Re_vapor
train3[:,16]  = Re_liquid
train3[:,17]  = Re
train3[:,18]  = Pe_vapor
train3[:,19]  = Pe_liquid
train3[:,20]  = Pe
train3[:,21]  = Su_vapor
train3[:,22]  = Su_liquid
train3[:,23]  = Su
train3[:,24]  = We_vapor
train3[:,25]  = We_liquid
train3[:,26]  = We
print(train3.shape)
datta3 = np.zeros((76,15)) 
datta3[:,0]   = datamatrix3[:,0]*1e3      # Pa
datta3[:,1]   = datamatrix3[:,1]          # K
datta3[:,2]   = datamatrix3[:,2]/1000     # Dh [m]
datta3[:,3]   = datamatrix3[:,3]          # °
datta3[:,4]   = datamatrix3[:,4]*1e4      # w/m2
datta3[:,5]   = datamatrix3[:,5]          # m/s
datta3[:,6]   = Bd
datta3[:,7]   = Fr 
datta3[:,8]  = Ga
datta3[:,9]  = Pr
datta3[:,10]  = Ka
datta3[:,11]  = Re
datta3[:,12]  = Pe
datta3[:,13]  = Su
datta3[:,14]  = We
```
